```python
import pandas as pd
from scipy.stats import pearsonr
```

```python
df=pd.read_csv('C:/Users/Home/Desktop/AIML/Day 7/Day-7/Assignment/general_data.csv')
```

```python
df.head()
```

```python
df.dropna()
```

```python
df.drop_duplicates()
```

```python
df.columns
```

```python
df.corr()
```

```python
df.Attrition.replace(to_replace = ('No','Yes'),value = (0,1),inplace = True)
```

```python
df[['Attrition','Age']].head(10)
```

```python
df.Gender.replace(to_replace = ('Male','Female'),value = (0,1),inplace = True)
```

```python
df
```

```python
stats,p=pearsonr(df.Attrition,df.Age)
print('Correlation test between Attrition and Age Variables\n')
print('r value is :',stats,'\nP value is :',p)
if stats < 0:
    print('\nNegetive Correlation between both variables')
else:
    print('\nPositive Correlation between both variables')
if p < 0.05:
    print('as p < 0.05, H0 is Rejected')
else:
    print('as P > 0.05, H0 is Accepted')
```
